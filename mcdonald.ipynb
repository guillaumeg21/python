```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='fr-FR')
```

```python
kw_list = ["/m/07gyp7"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='FR', gprop='')
df = pytrends.get_historical_interest(kw_list, year_start=2021, month_start=1, day_start=1, hour_start=0, year_end=2021, month_end=12, day_end=31, hour_end=23, cat=0, geo='', gprop='', sleep=1)
df.info()
```

L'index est bien Datetime.

```python
df=df['/m/07gyp7']
df.head()
```

```python
df.groupby(df.index.hour).mean().plot(figsize=(16,10))
```

```python
df.groupby(df.index.dayofweek).mean().plot(figsize=(16,10))
```

```python
df.groupby(df.index.day).mean().plot(figsize=(16,10))
```
